```python
#this code is being used to test QR code generation/reading capabilities
#using python coding

import qrcode
```

```python
#make a sample QR code to play with
sampleText = 'QR codes are great!'
img = qrcode.make(sampleText)
#save it to file and show it
img.save("Sample_QR.png")
img.show()
```

```python
import qrtools
```

```python
qr = qrtools.QR()
qr.decode('Sample_QR.png')
print qr.data
```
